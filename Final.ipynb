```
import os
import numpy as np
import pandas as pd
import cv2
import pickle
from keras.preprocessing import image
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
from skimage.filters import sobel
from scipy.ndimage import variance

print('All required libraries loaded...')
```

```
input_size = (512, 512)

X = []
y = []
y_pred = []
threshold = 0.0019
```

```
t = []
t1 = []
```

```
def variance_of_sobel(image):
  edge = sobel(image)
  return variance(edge)
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Undistorted/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size=(192,192))
        gray = cv2.cvtColor(np.asarray(img), cv2.COLOR_BGR2GRAY)
        fm = variance_of_sobel(gray)
        t1.append(fm)
        
        if fm < threshold:
            y_pred.append(1)
        else:
            y_pred.append(0)        

    else:
        print(filename, 'Not an image')
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Artificially-Blurred/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size=(192,192))
        gray = cv2.cvtColor(np.asarray(img), cv2.COLOR_BGR2GRAY)
        fm = variance_of_sobel(gray)
        t.append(fm)

        if fm < threshold:
            y_pred.append(1)
        else:
            y_pred.append(0)
        t.append(fm)
      
    else:
        print(filename, 'Not an image')
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Naturally-Blurred/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size=(192,192))
        gray = cv2.cvtColor(np.asarray(img), cv2.COLOR_BGR2GRAY)
        fm = variance_of_sobel(gray)
        
        t.append(fm)
        if fm < threshold:
            y_pred.append(1)
        else:
            y_pred.append(0)
        t.append(fm)
    
    else:
        print(filename, 'Not an image')
```

```
with open('/content/drive/My Drive/CloudSEK/y_train.pkl', 'rb') as picklefile:
    y_train = pickle.load(picklefile)
```

```
accuracy = accuracy_score(y_train, y_pred)
print("accuracy: {:.2f}%".format(accuracy * 100))
```

```
digBSet = pd.read_excel('/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/EvaluationSet/DigitalBlurSet.xlsx')
nBSet = pd.read_excel('/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/EvaluationSet/NaturalBlurSet.xlsx')
```

```
digBSet['MyDigital Blur'] = digBSet['MyDigital Blur'].apply(lambda x : x.strip())
digBSet = digBSet.rename(columns={"Unnamed: 1": "Blur Label"})
nBSet['Image Name'] = nBSet['Image Name'].apply(lambda x : x.strip())
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/EvaluationSet/DigitalBlurSet/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size= input_size)
        blur = digBSet[digBSet['MyDigital Blur'] == filename].iloc[0]['Blur Label']
        gray = cv2.cvtColor(np.asarray(img), cv2.COLOR_BGR2GRAY)
        fm = variance_of_sobel(gray)

        if fm < threshold:
            y_pred.append(1)
        else:
            y_pred.append(0)

        if blur == 1:
            y_test.append(1)
            t_blur.append(fm)
        else:
            y_test.append(0)
            t_nblur.append(fm)

    else:
        print(filename, 'not a pic')

print("Artificially Blurred Evaluated...")
```
