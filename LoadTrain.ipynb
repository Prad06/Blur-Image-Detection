```
import numpy as np
import pandas as pd
import os
import pickle
import cv2
from keras.preprocessing import image

print('All required libraries loaded...')
```

```
X_train = []
y_train = []

input_size = (96, 96)
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Undistorted/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size = input_size)
        X_train.append((1/255)*np.asarray(img))
        y_train.append(0)
    
    else:
        print(filename, 'Not an image')

print("Train Set: Undistorted Loaded...")
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Artificially-Blurred/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size=input_size)
        X_train.append((1/255)*np.asarray(img))
        y_train.append(1)
    
    else:
        print(filename, 'Not an image')

print("Train Set: Artificially Blurred Loaded...")
```

```
path = '/content/drive/My Drive/CloudSEK/CERTH_ImageBlurDataset/TrainingSet/Naturally-Blurred/'
```

```
for filename in os.listdir(path):
    if filename != '.DS_Store':
        imagepath = path + filename
        img = image.load_img(imagepath, target_size=input_size)
        X_train.append((1/255)*np.asarray(img))
        y_train.append(1)
    
    else:
        print(filename, 'Not an image')

print("Train Set: Naturally Blurred loaded...")
```

```
with open('X_train.pkl', 'wb') as picklefile:
    pickle.dump(X_train, picklefile)
with open('y_train.pkl', 'wb') as picklefile:
    pickle.dump(y_train, picklefile)

print("Train Set Loaded!!")
```
